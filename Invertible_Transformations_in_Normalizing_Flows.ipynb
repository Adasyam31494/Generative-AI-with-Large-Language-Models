```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x1 = np.linspace(1, 4, 100)
x2 = np.linspace(0, 2, 100)
X1, X2 = np.meshgrid(x1, x2)

P_original = ((X1 - 1) * X2) / 9
```

```python
fig1 = plt.figure(figsize=(10, 6))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X1, X2, P_original, cmap='viridis')
ax1.set_title('Original Distribution: $p(x_1, x_2) = \\frac{(x_1 - 1)x_2}{9}$')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$p(x_1, x_2)$')
plt.tight_layout()
plt.savefig("figure1_original_distribution.png")
plt.show()
```

```python
z1 = np.linspace(0, 1, 100)
z2 = np.linspace(0, 1, 100)
Z1, Z2 = np.meshgrid(z1, z2)

P_transformed = 4 * Z1 * Z2
```

```python
fig2 = plt.figure(figsize=(10, 6))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.plot_surface(Z1, Z2, P_transformed, cmap='plasma')
ax2.set_title('Transformed Distribution: $p(z_1, z_2) = 4z_1z_2$')
ax2.set_xlabel('$z_1$')
ax2.set_ylabel('$z_2$')
ax2.set_zlabel('$p(z_1, z_2)$')
plt.tight_layout()
plt.savefig("figure2_transformed_distribution.png")
plt.show()
```
