```python
import os
import zipfile
import math
import random
import numpy as np
from pathlib import Path
from tqdm import tqdm

import music21 as m21
import pretty_midi as pm

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras import mixed_precision

# Optional plotting used later for figures
import matplotlib.pyplot as plt

mixed_precision.set_global_policy("mixed_float16")
tf.config.optimizer.set_jit(True)

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

BASE_DIR = Path("/content")
DATA_DIR = BASE_DIR / "data" / "bach_cello_suites"
DATA_DIR.mkdir(parents=True, exist_ok=True)
OUT_DIR = BASE_DIR / "output"
OUT_DIR.mkdir(parents=True, exist_ok=True)

print("TensorFlow:", tf.__version__)
print("Mixed precision policy:", mixed_precision.global_policy())
print("Data dir:", DATA_DIR)
print("Output dir:", OUT_DIR)
```

```python
import urllib.request

ZIP_URL = "https://www.mutopiaproject.org/ftp/BachJS/BWV1007/bwv1007/bwv1007-mids.zip"
ZIP_PATH = DATA_DIR / "bwv1007-mids.zip"

def download_file(url, dest):
    if dest.exists() and dest.stat().st_size > 0:
        print("Already downloaded:", dest.name)
        return
    print("Downloading:", url)
    urllib.request.urlretrieve(url, dest)
    print("Saved to:", dest)

def extract_zip(zip_path, target_dir):
    with zipfile.ZipFile(zip_path, "r") as zf:
        zf.extractall(target_dir)
    print("Extracted to:", target_dir)

download_file(ZIP_URL, ZIP_PATH)
extract_zip(ZIP_PATH, DATA_DIR)

midi_files = sorted([p for p in DATA_DIR.glob("**/*.mid*")])
print("Found MIDI files:", len(midi_files))
for p in midi_files[:10]:
    print(" -", p.name)
assert len(midi_files) > 0, "No MIDI files found — check the download link."
```

```python
TICKS_PER_QUARTER = 4

def ql_to_ticks(quarter_length):
    ticks = int(max(1, round(quarter_length * TICKS_PER_QUARTER)))
    return int(min(ticks, 32))

def midi_to_note_dur_tokens(midi_path):
    s = m21.converter.parse(str(midi_path))
    part = s.parts[0] if s.parts else s.flat

    tokens = []
    for n in part.recurse().notesAndRests:
        if n.isNote:
            pitch = n.pitch.midi
            dur_ticks = ql_to_ticks(n.duration.quarterLength)
            tokens.append(f"n_{pitch}")
            tokens.append(f"d_{dur_ticks}")
        elif n.isRest:
            dur_ticks = ql_to_ticks(n.duration.quarterLength)
            tokens.append("n_rest")
            tokens.append(f"d_{dur_ticks}")
    return tokens

BOS, EOS, PAD = "[BOS]", "[EOS]", "[PAD]"
all_pieces = []

for mf in midi_files:
    try:
        tks = midi_to_note_dur_tokens(mf)
        if len(tks) >= 4:
            seq = [BOS] + tks + [EOS]
            all_pieces.append(seq)
    except Exception as e:
        print(f"Skipping {mf.name}: {e}")

print(f"Total pieces parsed: {len(all_pieces)}")
print("Example tokens (first 40):", all_pieces[0][:40])
```

```python
vocab = {PAD, BOS, EOS}
for seq in all_pieces:
    vocab.update(seq)

token2id = {tok: i for i, tok in enumerate(sorted(vocab))}
id2token = {i: tok for tok, i in token2id.items()}
vocab_size = len(token2id)

note_token_ids = sorted([token2id[t] for t in token2id if str(t).startswith("n_")])
dur_token_ids  = sorted([token2id[t] for t in token2id if str(t).startswith("d_")])

print("Vocab size:", vocab_size)
print("Note tokens:", len(note_token_ids), " Duration tokens:", len(dur_token_ids))

encoded_stream = []
for seq in all_pieces:
    encoded_stream.extend([token2id[t] for t in seq])

print("Encoded stream length:", len(encoded_stream))
```

```python
SEQ_LEN = 128
STEP = 8
BATCH_SIZE = 32

def make_windows(int_stream, seq_len=SEQ_LEN, step=STEP):
    X, y = [], []
    for i in range(0, len(int_stream) - seq_len - 1, step):
        X.append(int_stream[i : i + seq_len])
        y.append(int_stream[i + 1 : i + seq_len + 1])
    return np.array(X, dtype=np.int32), np.array(y, dtype=np.int32)

X, y = make_windows(encoded_stream, SEQ_LEN, STEP)
print("Dataset shapes:", X.shape, y.shape)

ds = tf.data.Dataset.from_tensor_slices((X, y)).cache()
ds = ds.shuffle(2048, seed=SEED, reshuffle_each_iteration=True)
ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
class PositionalEmbedding(layers.Layer):
    def __init__(self, vocab_size, d_model, max_len=2048):
        super().__init__()
        self.token_emb = layers.Embedding(vocab_size, d_model)
        self.pos_emb   = layers.Embedding(max_len, d_model)
    def call(self, x):
        positions = tf.range(start=0, limit=tf.shape(x)[1], delta=1)
        return self.token_emb(x) + self.pos_emb(positions)

class TransformerBlock(layers.Layer):
    def __init__(self, d_model, num_heads, d_ff, dropout=0.1):
        super().__init__()
        self.attn = layers.MultiHeadAttention(
            num_heads=num_heads,
            key_dim=d_model // num_heads,
            dropout=dropout
        )
        self.ffn  = tf.keras.Sequential([
            layers.Dense(d_ff, activation="gelu"),
            layers.Dense(d_model),
        ])
        self.norm1 = layers.LayerNormalization(epsilon=1e-5)
        self.norm2 = layers.LayerNormalization(epsilon=1e-5)
        self.drop1 = layers.Dropout(dropout)
        self.drop2 = layers.Dropout(dropout)

    def call(self, x, training=False):
        attn_out = self.attn(x, x, use_causal_mask=True, training=training)
        x = self.norm1(x + self.drop1(attn_out, training=training))
        ffn_out = self.ffn(x)
        x = self.norm2(x + self.drop2(ffn_out, training=training))
        return x

def build_model(vocab_size, d_model=192, num_heads=3, d_ff=384, num_layers=3, max_len=SEQ_LEN, dropout=0.1):
    inputs = layers.Input(shape=(max_len,), dtype=tf.int32)
    x = PositionalEmbedding(vocab_size, d_model, max_len=max_len)(inputs)
    x = layers.Dropout(dropout)(x)
    for _ in range(num_layers):
        x = TransformerBlock(d_model, num_heads, d_ff, dropout=dropout)(x)
    logits = layers.Dense(vocab_size, dtype="float32")(x)
    model = models.Model(inputs, logits, name="fast_transformer_lm")
    return model

model = build_model(vocab_size)
model.summary()

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
opt = tf.keras.optimizers.Adam(learning_rate=3e-4)
model.compile(optimizer=opt, loss=loss_fn, metrics=["sparse_categorical_accuracy"], run_eagerly=False)
```

```python
EPOCHS = 8

callbacks = [
    tf.keras.callbacks.ReduceLROnPlateau(monitor="loss", factor=0.5, patience=1, min_lr=1e-5, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor="loss", patience=2, restore_best_weights=True, min_delta=1e-3)
]

history = model.fit(ds, epochs=EPOCHS, callbacks=callbacks, verbose=1)
```

```python
def sample_from_logits(logits, allowed_ids, temperature=1.0):
    logits = tf.convert_to_tensor(logits, dtype=tf.float32)
    mask = tf.ones_like(logits) * -1e9
    idxs = tf.reshape(tf.constant(allowed_ids, dtype=tf.int32), (-1, 1))
    updates = tf.gather(logits, allowed_ids)
    allowed_logits = tf.tensor_scatter_nd_update(mask, indices=idxs, updates=updates)

    if temperature is None or temperature <= 0:
        temperature = 1.0
    allowed_logits = allowed_logits / temperature

    probs = tf.nn.softmax(allowed_logits)
    eps = tf.constant(1e-8, dtype=tf.float32)
    next_id = tf.random.categorical(tf.math.log(tf.expand_dims(probs + eps, 0)), 1)[0, 0].numpy().item()
    return next_id

def generate_tokens(model, start_note="n_60", start_dur="d_4", max_notes=80, temperature=1.0):
    prompt = [token2id[BOS], token2id.get(start_note, token2id["n_60"]), token2id.get(start_dur, token2id["d_4"])]
    seq = prompt[:]

    while len(seq) < SEQ_LEN:
        seq.insert(0, token2id[PAD])

    produced_pairs = 1
    expect_note = True

    eos_id = token2id[EOS]

    while produced_pairs < max_notes:
        x = np.array(seq[-SEQ_LEN:])[None, ...]
        logits = model.predict(x, verbose=0)[0, -1]

        allowed = note_token_ids if expect_note else dur_token_ids

        if not expect_note and produced_pairs >= 8:
            allowed = allowed + [eos_id]

        next_id = sample_from_logits(logits, allowed, temperature=temperature)
        seq.append(next_id)

        expect_note = not expect_note

        if expect_note:
            produced_pairs += 1

        if next_id == eos_id:
            break

    return seq

def tokens_to_midi(token_ids, out_path):
    s = m21.stream.Stream()
    inst = m21.instrument.Violoncello()
    s.insert(0, inst)

    toks = [id2token.get(i, PAD) for i in token_ids]

    i = 0
    while i < len(toks) and toks[i] in (PAD, BOS):
        i += 1

    while i < len(toks) and not str(toks[i]).startswith("n_"):
        i += 1

    while i + 1 < len(toks):
        nt, dt = toks[i], toks[i+1]
        if not (str(nt).startswith("n_") and str(dt).startswith("d_")):
            i += 1
            continue
        i += 2

        if nt == "n_rest":
            ql = int(dt.split("_")[1]) / TICKS_PER_QUARTER
            r = m21.note.Rest()
            r.duration = m21.duration.Duration(quarterLength=ql)
            s.append(r)
        else:
            pitch = int(nt.split("_")[1])
            ql = int(dt.split("_")[1]) / TICKS_PER_QUARTER
            n = m21.note.Note(pitch)
            n.duration = m21.duration.Duration(quarterLength=ql)
            n.volume.velocity = 80
            s.append(n)

    out_path = Path(out_path)
    out_path.parent.mkdir(parents=True, exist_ok=True)
    s.write("midi", fp=str(out_path))
    return str(out_path)
```

```python
SEED_NOTE = "n_60"
SEED_DUR  = "d_4"
TEMPERATURE = 1.0
MAX_NOTES = 80

gen_ids = generate_tokens(
    model,
    start_note=SEED_NOTE,
    start_dur=SEED_DUR,
    max_notes=MAX_NOTES,
    temperature=TEMPERATURE
)

out_midi_path = OUT_DIR / "generated_bach_cello_like.mid"
saved_path = tokens_to_midi(gen_ids, out_midi_path)
print("Saved generated MIDI to:", saved_path)

midi_obj = pm.PrettyMIDI(str(saved_path))
print("Generated MIDI length (seconds):", round(midi_obj.get_end_time(), 2))
```

```python
import pretty_midi as pm
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

FIG_DIR = Path("/content/figures")
FIG_DIR.mkdir(parents=True, exist_ok=True)
fig2_path = FIG_DIR / "figure2_generated_pianoroll.png"

def plot_piano_roll_sum(midi_path, fs=100, title="Piano Roll"):
    midi = pm.PrettyMIDI(str(midi_path))
    roll = midi.get_piano_roll(fs=fs)
    roll = np.clip(roll, 0, 127)

    plt.figure(figsize=(12, 4))
    plt.imshow(roll, aspect="auto", origin="lower", interpolation="nearest")
    plt.xlabel(f"Time (frames @ {fs} fps)")
    plt.ylabel("MIDI Pitch")
    plt.title(title)
    plt.tight_layout()
    return roll

roll = plot_piano_roll_sum(out_midi_path, fs=100, title="Generated MIDI — Piano Roll")
plt.savefig(fig2_path, dpi=200)
plt.show()

print("Saved:", fig2_path, " | roll shape:", roll.shape)
```

```python
import numpy as np
import pretty_midi as pm
import matplotlib.pyplot as plt
from pathlib import Path

FIG_DIR = Path("/content/figures")
FIG_DIR.mkdir(parents=True, exist_ok=True)

ORIGINAL_MIDI_PATH = None

fallback_midi = midi_files[0] if len(midi_files) > 0 else None

def note_activity_curve(midi_path, fs=200):
    """Return a 1D normalized activity curve using summed piano-roll energy."""
    midi = pm.PrettyMIDI(str(midi_path))
    roll = midi.get_piano_roll(fs=fs)
    activity = roll.sum(axis=0)
    if activity.max() > 0:
        activity = activity / activity.max()
    return activity

def plot_activity(activity, title, save_path):
    plt.figure(figsize=(12, 3))
    plt.plot(activity)
    plt.xlabel("Time (frames)")
    plt.ylabel("Normalized Activity")
    plt.title(title)
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.show()
    print("Saved:", save_path)

act_opt = note_activity_curve(out_midi_path, fs=200)
fig3a_path = FIG_DIR / "figure3a_optimized_activity.png"
plot_activity(act_opt, "Optimized Model Output — Activity Over Time", fig3a_path)

if ORIGINAL_MIDI_PATH is not None and Path(ORIGINAL_MIDI_PATH).exists():
    comp_path = ORIGINAL_MIDI_PATH
    comp_title = "Original (Reference) Model Output — Activity Over Time"
    fig3b_path = FIG_DIR / "figure3b_original_activity.png"
else:
    comp_path = fallback_midi
    comp_title = "Source Bach MIDI (Dataset) — Activity Over Time"
    fig3b_path = FIG_DIR / "figure3b_dataset_activity.png"

if comp_path is not None:
    act_ref = note_activity_curve(comp_path, fs=200)
    plot_activity(act_ref, comp_title, fig3b_path)
else:
    print("No comparison MIDI available. Set ORIGINAL_MIDI_PATH or ensure dataset MIDI is present.")
```
